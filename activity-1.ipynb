### Importing Libraries

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

### Loading Images

```python
img_filenames = os.listdir('data')
imgs = [mpimg.imread(os.path.join('data', img_filename)) for img_filename in img_filenames]
```

### Creating Sub-plots

```python
# Create a grid of subplots sized to the number of images
n = len(imgs)
ncols = 2
nrows = int(np.ceil(n / ncols))
fig, axes = plt.subplots(nrows, ncols, figsize=(6, 6), dpi=150)

# Flatten axes to 1D array for easy indexing
axes = np.array(axes).ravel()

# Specify labels
labels = ['coast', 'beach', 'building', 'city at night']

# Plot images (only as many as we have axes for)
for i, img in enumerate(imgs):
    axes[i].imshow(img)
    axes[i].set_xticks([])
    axes[i].set_yticks([])
    if i < len(labels):
        axes[i].set_xlabel(labels[i])

# Turn off any unused axes
for j in range(len(imgs), axes.size):
    axes[j].axis('off')
```
