### Importing Libraries

```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from PIL import Image
```

### Loading Image

```python
im = Image.open("data/colourSpaces.jpeg")
```

### Load Image and Create 2D array of pixels

```python
px = im.load()
```

### Create 3D array

```python
ax = plt.axes(projection='3d')

# Question 1) Create empty matrices for each coordinate as well as color
x = []
y = []
z = []
c = []

# Question 2) Normalize the RGB values between 0 and 1 when providing colors
for row in range(0, im.height):
    for col in range(0, im.width):
        pix = px[col, row]                      # get pixel RGB values (tuple)
        newCol = (pix[0]/255, pix[1]/255, pix[2]/255)  # normalize RGB to [0,1]

        # avoid duplicates for efficiency
        if newCol not in c:
            x.append(pix[0])
            y.append(pix[1])
            z.append(pix[2])
            c.append(newCol)

# Plot the 3D scatter of RGB values
ax.scatter(x, y, z, c=c)

plt.show()
```
